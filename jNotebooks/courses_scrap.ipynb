```python
import json
import requests
from bs4 import BeautifulSoup
```

```python
course_url = "https://ineuron.ai/courses"
```

```python
res = requests.get(course_url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
courses = soup.find("script", {"id": "__NEXT_DATA__"})
```

```python
courses_json = json.loads(courses.text)
```

```python
courses_json['props']['pageProps'].keys()
```

```python
courses_json['props']['pageProps']['initialState'].keys()
```

```python
courses_json['props']['pageProps']['initialState']['init'].keys()
```

```python
courses_list = list(courses_json['props']['pageProps']['initialState']['init']['courses'].keys())
```

```python
courses_list[:5]
```

```python
base_url ="https://ineuron.ai/course/"
for cl in courses_list:
    tl = cl.replace(" ", "-")
    tl = base_url+tl
    # res = requests.get(tl)
    # print(tl)
    # print(res)
```
